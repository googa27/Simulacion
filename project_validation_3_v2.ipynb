```python
curr_sheet = "N_Trenes"
raw_data = dfs_data[curr_sheet]
iid_data = dfs_data[curr_sheet].values.flatten()
n = len(iid_data)

t_score = st.t.ppf(0.975, n - 1)
chi_score_high = st.chi2.ppf(0.975, n-1)
chi_score_low = st.chi2.ppf(0.025, n-1)
sig = iid_data.std()
var = iid_data.var(ddof=1)
delta = t_score*sig/np.sqrt(n)
mu = iid_data.mean()

print(delta)
print("Standard deviation estimator:", sig)
print("Variance estimator: ", var)
print("Mean estimator :", mu)

var_low = (n-1)*var/chi_score_high
var_high = (n-1)*var/chi_score_low

print("Confidence Interval for mean: (" + "{:.5f}".format(mu - delta) + ", " + "{:.5f}".format(mu + delta) + ")")
print("Confidence Interval for var: (" + "{:.5f}".format(var_low) + ", " + "{:.5f}".format(var_high) + ")")

##################### REAL PARAMETERS ###########################

probs = np.array([0.05, 0.2, 0.45, 0.3])
vals = np.arange(4) + 1

real_mean = probs.dot(vals)
real_var = probs.dot((vals - real_mean)**2)
real_std = np.sqrt(real_var)

print("Real mean: ", real_mean)
print("Real standard deviation: ", real_std)
print("Real variance: ", real_var)
```
