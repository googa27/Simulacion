```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def leer_hoja(file, sheet_name):
    assert(sheet_name in {"Bodega_Puerto", "Bodega_Saladillo"})

    return pd.read_excel(file, index_col=0, header = 1, sheet_name=sheet_name)

def leer_excel(file):
    return leer_hoja(file, "Bodega_Saladillo"), leer_hoja(file, "Bodega_Puerto")

def to_series(df):
    return pd.concat([df.iloc[:, i] for i in range(df.shape[1])], keys = np.arange(df.shape[1]))

def series_df(df_list, col_names):
#     series = [to_series(df).values for df in df_list]
    series = [to_series(df) for df in df_list]
    
    out = pd.concat(series, axis=1)
#     out = pd.DataFrame({col_name: series[i] for i, col_name in enumerate(col_names)})
    out.columns = col_names
    return out
```
